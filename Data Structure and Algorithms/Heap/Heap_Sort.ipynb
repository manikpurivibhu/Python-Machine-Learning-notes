# Heap Sort

Heapsort is a comparison based sorting technique <b>based on a Binary Heap data structure</b>. <br><br>

We take in an Array and <b>create a Max-Heap</b> data structure so that the largest element is always at the root (top).<br>

Next we <b>delete the root node and place it in the space freed</b> because of this delete operation. This puts the largest element at the end of the array.<br>

We <b>repeat</b> the above step until all the elements are deleted i.e. the array is sorted


```python
def swap(array,i, j):
    array[i], array[j] = array[j], array[i]
    
def heapifyDown(array, parentIndex, heap_size):
    while True:
        leftChildIndex  = 2*parentIndex + 1
        rightChildIndex = 2*parentIndex + 2
        
        if max(leftChildIndex, rightChildIndex) < heap_size:
            
            if array[parentIndex] >= max(array[leftChildIndex], array[rightChildIndex]):
                break
                
            elif array[leftChildIndex] > array[rightChildIndex]:   
                swap(array, parentIndex, leftChildIndex)
                parentIndex = leftChildIndex

            else:
                swap(array, parentIndex, rightChildIndex)
                parentIndex = rightChildIndex
        
        #if only left child exists
        elif leftChildIndex < heap_size:
            if array[parentIndex] < array[leftChildIndex]:
                swap(array, parentIndex, leftChildIndex)
                parentIndex = leftChildIndex
            else:
                break
                
        elif rightChildIndex < heap_size:
            if array[parentIndex] < array[rightChildIndex]:
                swap(array, parentIndex, rightChildIndex)
                parentIndex = rightChildIndex
            else:
                break
                
        else:
            break
```

```python
def heap_sort(array):
    for i in range((len(array)-2)//2, -1, -1):
        heapifyDown(array, i, len(array))
        
    for end in range(len(array)-1, 0, -1):
        swap(array, 0, end)
        heapifyDown(array, 0, end)
```

```python
##Test:

sample_array = [5, 16, 8, 14, 20, 1, 26]
print('Array before sorting :', sample_array)
heap_sort(sample_array)
print('Array after sorting  :', sample_array)
```
