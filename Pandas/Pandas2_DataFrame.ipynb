<b>.isin()<b>

this method takes in a list of values and returns (T/F) if those val.s are in the df['col']

```python
c1 = df['education'] == 'Masters'
c2 = df['education'] == '11th'
c3 = df['education'] == '12th'
del_temp = df[c1 | c2 | c3]
print(del_temp.head(10))
```

```python
del_temp2 = df[df['education'].isin(['Masters', '11th', '12th'])]
print(del_temp2)
```

# .loc() and .iloc() method

these mehods are used to get individual rows of information from DataFrame
<br><br>.loc() to get get rows of information based on labelled indexing
<br>.iloc() to get get rows of information based on numerixcindexing

<br><br><br>df.loc['labelled row] to return row info of passed row
<br><br>df.loc['labelled row1', 'labeeled row2', ..] to return row info of passed rows
<br><br>df.loc['labelled row']['col']  to return 'col' info of passed row
<br><br>df.loc['labelled row']['col', 'col2',..]  to return info of passed rows and col.s
<br><br><br>
<br><br><br>df.iloc['index'] to return row of specified index
<br><br>df.iloc[index1, index2, ..] to return row info of passed rows
<br><br>df.iloc[index]['col']  to return 'col' info of passed row
<br><br>df.iloc[index1,index2,..]['col', 'col2',..]  to return info of passed rows and col.s

# .apply()

apply(func) allows user to apply user defined function to every el in row or entire dataset
<br><br> to apply to single col:       df['colname'].apply(userfunc)
<br> to apply to entire dataser:   df.apply(func = userfunc, axis = 'columns'))
<br><br>here, "columns" is keyword
<br>ps: you obviously need to define the user defined function fitrst before applying it

```python
#in dataset salary is either <= 50 K or >= 0 K 
def encodesalary(salary):
    if salary == " <=50K":  return '1'
    else:  return '0'
```

```python
df['salary'].apply(encodesalary)
```

```python
df['encoded_Salary'] = df['salary'].apply(encodesalary)
```

```python
df.columns
```

```python
def desc(row):
    age = row[0]
    sex = row[9]
    sal = row[14]
    return (sex + " earning "+ sal + "$")
```

```python
df.apply(func = desc, axis = 'columns')
```

```python
df['money'] = df.apply(func = desc, axis = 'columns')
```

```python
df.info()
```
