```python
names = ['age', 'workclass','fnlwgt','education','edu-num', 'marital-status','occupation','relationship',
            'race','sex','capital-gain','capital-loss','hours_per_week','native-country', 'salary']
df = pd.read_csv('adult.data.csv', header = None, names = names)
```

```python
df.head()
```

```python
df['sex'].unique()
```

```python
df['sex'].value_counts()
```

```python
df['sex'].value_counts().plot(kind = 'bar')
```

```python
df['sex'].value_counts().plot(kind = 'barh')
```

```python
df['education'].value_counts().plot(kind = 'barh')
```

```python
 df['education'].value_counts().plot(kind = 'pie')
```

```python
df['age'].plot(kind = 'hist')
```

```python
df['age'].plot(kind = 'hist', bins = 40)
#If bins is an integer, it defines the number of equal-width bins in the range.
#If bins is a sequence, it defines the bin edges, including the left edge of the first bin and the right edge of the last bin
#If bins is a string, it is one of the binning strategies
```
