```python
## 1D DFT, FFT, Inverse FFT
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1D FFT

t = np.arange(128)
print(t)

sp = np.fft.fft(np.sin(t))
freq = np.fft.fftfreq(t.shape[-1])
```

```python
print(sp, freq)
```

```python
plt.subplot(2,1,1)
plt.title('Real Part')
plt.plot(freq, sp.real)

plt.subplots_adjust(hspace = 0.5)

plt.show()

plt.subplot(2,1,1)
plt.title('Imaginary Part')
plt.plot(freq, sp.imag)

plt.show()
```

```python
#Inverse 1D DFT

t = np.arange(500)
n = np.zeros((500,),dtype = 'complex')

print(t,n)

```

```python

t = np.arange(500)
n = np.zeros((500,),dtype = 'complex')

#Randomizing some el.s of array
n[40:80] = np.exp(1j*np.random.uniform(0,2*np.pi, (40,)))
s = np.fft.ifft(n)  

#print(t,n)

plt.subplot(2,1,1)
plt.title('Real Part')
plt.plot(t, s.real)

plt.subplots_adjust(hspace = 0.5)

plt.show()

plt.subplot(2,1,1)
plt.title('Imaginary Part')
plt.plot(t, s.imag)

plt.show()   
```

```python
#Plotting them in the same graph instead of subplotting them
plt.plot(t,s.real)
plt.title("IFFT")
plt.plot(t,s.imag)
plt.show()
```
