```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(5)
print(x)
```

```python
y = x
print(y)
```

```python
#plot([x], y, [fmt], *, data=None, **kwargs)
#plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
#fmt is Format strings: just an abbreviation for quickly setting basic line properties
#ex : 'ro' for circles
plt.plot(x,y, 'o')
plt.show()
```

```python
plt.plot(x,y, '--o')
plt.plot(x,-y, '-ro')
plt.show()
# o : circles ; -o : line joining pt.s ; o-- : dotted line
```

```python
plt.plot(x,y, '--o')
plt.plot(x,-y, '-ro')
plt.title('y=x and  y=-x')
plt.show()
#plt.title('str) to display a title 'str' aloongwith the plot
```

```python
N = 11
#np.linspace(lower bound, upper bound, no of steps) can be used instead of np.arange
x = np.linspace(0, 10, N)
print(x)
y = x 
```

```python

plt.plot(x,y, '--o')
plt.plot(x,-y, '-ro')
plt.title('y=x and  y=-x')
plt.axis('off')
#plt.axis('off') hides the axes in display
plt.show()
```

```python
#Create logspace
y = np.logspace(0.1, 1, N)
print(y)
plt.plot(x,y)
plt.show()
```

```python
#Geometric Progression
y = np.geomspace(0.1, 1000, N)
print(y)
plt.plot(x,y,'--o')
plt.show()
```
