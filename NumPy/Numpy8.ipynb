```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Compute FFT of 2D matrix

a = np.mgrid[:3, :3][0]
print(a)
```

```python
print(np.fft.fft2(a))
```

```python
#n-dimensional FFT
```

```python
a = np.mgrid[:3,:3,:3][0]
print(a)
```

```python
print(np.fft.fftn(a, axes = (1,2)))
```

```python
print(np.fft.fftn(a, axes = (0,1)))
```

```python
[X, Y] = np.meshgrid( 2 * np.pi * np.arange(200) / 12, 2 * np.pi * np.arange(200)/34)
plt.plot(X,Y)
plt.show()
```

```python
n = np.sin(X) + np.cos(Y) + np.random.uniform(0,1, X.shape)
plt.imshow(n)
plt.show()
```

```python
plt.imshow(np.fft.fftn(n).astype(np.uint8))
plt.show()
```

```python
#Inverse N-D FFT
```

```python
n = np.zeros((200,200),dtype = 'complex')
n[60:80, 20:40] = np.exp(1j*np.random.uniform(0, 2*np.pi, (20,20)))

plt.imshow(np.fft.ifftn(n).real, cmap = 'hot')
plt.show()
```
