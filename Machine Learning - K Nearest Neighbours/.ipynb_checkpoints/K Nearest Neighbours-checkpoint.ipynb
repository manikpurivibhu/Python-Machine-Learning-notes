# K Nearest Neighbours



**K-nearest neighbors (KNN)** is a Supervised Machine Learning algorithm that can be used to solve both classification and regression problems. <br>
KNN is as an algorithm that is inspired from real life: people tend to be effected by the people around them.  There are ofcourse exceptions. KNN works similarly. <br>
KNN is also known as an **instance-based model** or a **lazy learner** because it doesnâ€™t construct an internal model.

The KNN algorithm assumes that similar things exist in close proximity. In other words, similar things are near to each other. <br>
"Birds of a feather flock together"

### Finding the "nearest neighbours"

The value of a data point is determined by the data points around it i.e. it's neighbours. <br>

There are many ways to calculate the similarity (or conversely, the dissimilarity) between data points, such as the Euclidean distance: $ \sqrt(\sum \limits _{i=1}^{n}(trainX_i - queryX_i)^2) $, Minkowski distance and more. <br>
The choice of distance measure (one might be preferable over others) is highly dependent on data types and the problem we are trying to solve


## KNN Algorithm

KNN works by finding the distances between a query and all the examples in the data (training set), selecting the specified number examples (K) closest to the query, then votes for the most frequent label (in the case of classification) or averages the labels (in the case of regression).

1. Receive an unclassified data and value of `K` (user defined)

2. Measure the distance (Euclidian, Manhattan, Minkowski or Weighted) from the new data to all others data that is already classified (training set)

3. Sort the calculated distances

4. Select "K nearest neighbours" i.e. K values from sorted distances

5. Takes mode of selected values in Classification problem or mean of selected values in Regression problem

6. Optimize the value of "K" for given data

### Choosing the right value for K

- K is small: The model is too specific and not generalized well. High Variance, Low Bias; results in Overfitting

- K is large: The model is too generalized and not a good predictor on both train and test sets. Low Variance, High Bias; results in Underfitting 

Generally, K gets decided based on the square root of the number of data points. It is also recommended to use an odd value for K to have a tie breaker in Classification problems.

Train using different values of K and compare the performance of each model. Use an error plot or accuracy plot to find the most favorable K value: Plot possible K values against error and find the K with minimum error. <br>


Scikit-learn provides GridSearchCV function that allows us to easily check multiple values for K. 


### KNN is lazy learner

KNN algorithm is known as an instance-based method or lazy learner because **it doesnâ€™t explicitly learn a model**. Unlike other Supervised Machine Learning algorithms like Linear Regression or Support Vector Machine, which learn a discriminative function from the training data (the underlying relationship among features) and implement that learning during prediction, KNN does not learn anything during training. It just "memorizes the training dataset" which are used as â€œknowledgeâ€ for the prediction phase. Simply put, all calculations are performed directly while making predictions. 

To summarize: An eager learner has a model fitting or training step. A lazy learner does not have a training phase.

Hence, **KNN is a non-parameteric algorithm**. <br>
Algorithms that do not make particular assumptions (ex. Linear Regression: $ y = \theta^T X $) about the kind of mapping function are known as non-parametric algorithms.

### Advantages and Disadvantages / pros and cons

**Pros**
- Simple and easy to interpret
- Does not make any assumption, so it can be implemented in non-linear tasks
- Works well on classification with multiple classes
- Works on both classification and regression tasks
- Can also be used to generate a (weak) Recommendation System

**Cons**
- Becomes very slow as the number of data points increases because the model needs to store all data points
- Not memory efficient
- Sensitive to outliers. Outliers also have a vote!


## KNN from scratch

**NOTE:** Following implementation are for understanding the K-Nearest neighbors algorithm and the mathematics that support it. <br>
In practice, use `scikit-learn` modules (discussed later in this notebook) or other standard libraries or frameworks to implement K-Nearest Neighbors. <br>
The actual results and calculations between following implementation and sklearn's implementation may vary

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
def most_common(lst):
    return max(set(lst), key=lst.count)

def average(lst):
    return np.mean(lst)

def euclidean(point, data):
    '''Euclidean distance between a point and training data '''
    return np.sqrt(np.sum((point - data)**2, axis=1))
```

```python
class KNearestNeighbors:
    
    def __init__(self, k=5, dist_metric=euclidean):
        self.k = k
        self.dist_metric = dist_metric
        
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def predict(self, X_test, problem='Classification'):
        '''
        for each test data, 
        calculate their distance from each training data,
        wrap the distance and their corresponding output (y_train) together and sort them,
        select the closest (first) K,
        predict based on mode or mean for Classification or Regression problem respectively
        '''
        neighbors = []
        for x in X_test:
            distances = self.dist_metric(x, self.X_train)
            y_sorted = [y for _, y in sorted(zip(distances, self.y_train))]
            neighbors.append(y_sorted[:self.k])
        if problem=='Classification':
            return list(map(most_common, neighbors))
        return list(map(average, neighbors))
    
    def evaluate(self, X_test, y_test, problem='Classification'):
        if problem=='Classification':
            y_pred = self.predict(X_test, problem='Classification')
            accuracy = sum(y_pred == y_test) / len(y_test)
            return round(accuracy*100, 2)
        else:
            y_pred = self.predict(X_test, problem='Regression')
            diff = y_test - y_pred
            sq = np.square(diff)
            avg = np.mean(sq)
            rmse = np.sqrt(avg)
            return round(rmse, 3)
```

```python
# Classification 

iris = datasets.load_iris()
X = iris['data']
y = iris['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler().fit(X_train)
X_train, X_test = scaler.transform(X_train), scaler.transform(X_test)

# Test knn model across varying ks
accuracies = []
ks = range(1, 30)
for k in ks:
    knn = KNearestNeighbors(k=k)
    knn.fit(X_train, y_train)
    accuracy = knn.evaluate(X_test, y_test, 'Classification')
    accuracies.append(accuracy)
    
# Visualize accuracy vs. k

fig, ax = plt.subplots()
ax.plot(ks, accuracies)
ax.set(xlabel="K",
       ylabel="Accuracy",
       title="Performance of KNN")
plt.show()
print('Optimal K :', accuracies.index(max(accuracies))+1)
```

```python
# Regression

from sklearn import datasets
boston = datasets.load_boston()

X, y = boston.data, boston.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler().fit(X_train)
X_train, X_test = scaler.transform(X_train), scaler.transform(X_test)

# Test knn model across varying ks
errors = []
ks = range(1, 30)
for k in ks:
    knn = KNearestNeighbors(k=k)
    knn.fit(X_train, y_train)
    rmse = knn.evaluate(X_test, y_test, 'Regression')
    errors.append(rmse)
    
# Visualize accuracy vs. k

fig, ax = plt.subplots()
ax.plot(ks, errors)
ax.set(xlabel="K",
       ylabel="Root Mean Square Error",
       title="Performance of KNN")
plt.show()
print('Optimal K :', accuracies.index(min(accuracies))+1)
```

## KNN using sklearn

### for Classification

```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# load data
df = pd.read_csv('KNNdataset.csv')
X, y = df.drop('custcat', axis=1), df['custcat']

# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.25, random_state=0)

# KNN using sklearn
from sklearn.neighbors import KNeighborsClassifier  
knn = KNeighborsClassifier(n_neighbors= 3,weights = 'distance' ,metric="euclidean")
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Evaluation Metrics
from sklearn.metrics import accuracy_score
print("KNN Accuracy for K=3:",accuracy_score(y_test, y_pred)*100)

from sklearn.model_selection import GridSearchCV
knn_grid = GridSearchCV(estimator=KNeighborsClassifier(), 
                        param_grid={'n_neighbors': np.arange(1, 30)}, cv=4 )
knn_grid.fit(X, y)
K = knn_grid.best_params_['n_neighbors']
print('Optimal K for dataset :', K)

# Visualize accuracy vs. K
Ks = np.arange(1, 30)
scores = knn_grid.cv_results_['mean_test_score']
fig, ax = plt.subplots()
ax.plot(Ks, scores)
ax.set(xlabel="K",
       ylabel="Accuracy",
       title="Performance of KNN")
plt.show()

# final model
classification_model = KNeighborsClassifier(n_neighbors= K, weights='distance', metric="euclidean")
classification_model.fit(X, y)
```

### for Regression

```python
# import and transform data
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
boston = datasets.load_boston()

X, y = boston.data, boston.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
scaler = StandardScaler().fit(X_train)
X_train, X_test = scaler.transform(X_train), scaler.transform(X_test)

# KNN using sklearn
from sklearn.neighbors import KNeighborsRegressor  

knn = KNeighborsRegressor(n_neighbors= 3,weights='distance', metric="euclidean")
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Evaluation Metrics
from sklearn.metrics import r2_score
print("KNN Accuracy for K=3:", r2_score(y_test, y_pred))

from sklearn.model_selection import GridSearchCV
knn_grid = GridSearchCV(estimator=KNeighborsRegressor(), 
                        param_grid={'n_neighbors': np.arange(1, 30)}, cv=5)
knn_grid.fit(X, y)
K = knn_grid.best_params_['n_neighbors']
print('Optimal K for dataset :', K)

# Visualize accuracy vs. K
Ks = np.arange(1, 30)
scores = knn_grid.cv_results_['mean_test_score']
fig, ax = plt.subplots()
ax.plot(Ks, scores)
ax.set(xlabel="K",
       ylabel="R2 score",
       title="Performance of KNN")
plt.show()

# final model
regression_model = KNeighborsRegressor(n_neighbors= K, weights='distance', metric="euclidean")
regression_model.fit(X, y)
```

### Recommender system using KNN

Since KNN hinges on the assumption that similar data points are close to each other, it can also be used to generate a (weak) Recommender System by predicting the K data points which are most similar (nearest) to the input data point

The k-nearest neighbors (KNN) algorithm is a simple, easy to understand and implement Supervised Machine Learaning algorithm since it requires only 2 parameters (K and distance function). It can b e used to solve both Regression and Classificatino problems, but has a major drawback of becoming significantly slows as the size of that data in use grows.
